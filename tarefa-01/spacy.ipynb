```python
import spacy
```

```python
nlp = spacy.load("pt_core_news_lg")
```

```python
print(type(nlp))
```

```python
documento = nlp("A caminhada de Beatriz Ferreira rumo a uma histórica medalha de ouro se iniciou nesta segunda-feira nas Olímpiadas de Paris 2024. E começou da melhor forma: com vitória por decisão unânime (5-0) sobre a americana Jajaira Gonzalez nas oitavas de final da categoria peso-leve (até 60kg).")

print(len(documento.vocab))
print(type(documento))
```

```python
for tokens in documento:
    print(tokens.text)
```

```python
print("Tokens: ", [token.text for token in documento])
print("Stop words: ", [token.text for token in documento if token.is_stop == True])
print("Alfa numéricos: ", [token.text for token in documento if token.is_alpha == True])
print("Maiusculas: ", [token.text for token in documento if token.is_upper == True])
print("Pontuação: ", [token.text for token in documento if token.is_punct == True])
print("Numero: ", [token.text for token in documento if token.like_num == True])
```

```python
print("Tokens: ", [token.text for token in documento])
print("Formato: ", [token.shape_ for token in documento])
```

```python
for token in documento:
    print(token.text, " - ", token.lemma_, " - ", token.pos_, " - ", token.tag_, " - ", token.dep_, " - ", token.shape_, " - ", token.is_alpha, " - ", token.is_stop)
```

```python
for ent in documento.ents:
    print(ent.text, " - " ,ent.label_)
```

```python
for stopword in documento:
    if stopword.is_stop:
        print(stopword.text)
```

```python
frase1 = "Não, vá embora"
frase2 = "Não vá embora"

print("Silimalidade: ", nlp(frase1).similarity(nlp(frase2)))
```

```python
from typing import Match
from spacy.matcher import Matcher

telefones = nlp("Meu telefone é (45) 12345678 e o outro é (46) - 987654321")

matcher = Matcher(nlp.vocab)
padrao = [{"ORTH": "("}, {"SHAPE": "dd"}, {"ORTH": ")"}, {"ORTH": "-", "OP":"?"}, {"IS_DIGIT": True}]
matcher.add("TELEFONE", [padrao])
mathces = matcher(telefones)
for match_id, start, end in mathces:
    print(telefones[start:end])
```
